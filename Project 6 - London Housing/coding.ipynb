# what is the max and min 'average_price' per year in england?

```python
df['year']=df.date.dt.year
```

```python
df_eng = df[df.area=='england']
```

```python
df_eng
```

```python
df_eng.groupby('year').average_price.max()
```

```python
df_eng.groupby('year').average_price.min()
```
